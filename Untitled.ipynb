```python
for classes in classes_folder:
    plant_folder = not_clean_dec + classes 
    print(plant_folder)
    plant_images = os.listdir(plant_folder)
    clean_plant_dic = clean_dic + classes
    os.mkdir(clean_plant_dic)
    for images in plant_images:
        image_dic = plant_folder + "\\" + images
        notclean_img = cv2.imread(image_dic) #reads image
        notclean_img = 
        data_min = np.min(notclean_img, axis=(1,2), keepdims=True)
        data_max = np.max(notclean_img, axis=(1,2), keepdims=True)
        scaled_data = 255  * ((notclean_img - data_min) / (data_max - data_min)) #scaled image 
        #write images
        cv2.imwrite(clean_plant_dic+"//"+images, scaled_data)
        
```
